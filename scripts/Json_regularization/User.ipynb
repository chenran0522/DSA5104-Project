```python
import json
import random
import string
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NationalNames.csv")
filtered_data = df[df["Count"] > 10000]
unique_names = filtered_data["Name"].unique()
```

```python
users = []
for i in range(50000):
    user_id = i+100001
    password = ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(8))  # 随机生成8位密码
    name = random.choice(unique_names)
    user_info = {"user_id": user_id, "password": password, "name": name}
    users.append(user_info)

with open("user_data.json", "w") as json_file:
    json.dump(users, json_file, indent=4)

print("User data saved to user_data.json")
```

```python
# type 1 image 0 - 131035
# type 2 books
# type 3 music
# type 4 games
# 生成随机score和comment
import json
import random
import numpy as np
import string

# 生成随机score和comment
def generate_random_score():
    mean = 5.0 
    std_deviation = 2.0 
    score = np.random.normal(mean, std_deviation)
    
    score = max(0.0, min(10.0, score))
    
    return round(score, 1)

def generate_random_comment():
    beginningss = ["有一说一","不吹不黑","实话实说","说到底",
    "放下所有偏见，",
    "开门见山地说，",
    "扯远了不说，",
    "一开始就得说，",
    "先不多说，",
    "说起这个话题，",
    "在我开始评论前，",
    "直奔主题，",
    "不废话了，",
    "不拐弯抹角，",
    "不扯皮，",
]
    
    beginnings = [
    "这部电影",
    "从我的角度看",
    "我觉得这部作品",
    "对我来说，这部电影",
    "作为一个观众，我认为这部电影",
    "对于这个电影，我要说",
    "谈到这部影片，我想提到",
    "观看了这部作品后，我认为",
    "作为一个影迷，我深深感受到",
    "不得不说，这部电影",
    "在这个电影中，",
    "当我看这部作品时，",
    "关于这部电影，",
    "每个人对于这部电影的看法都不同，",
    "让我来说说这部电影"
    ]

    middles = [
        "真的很棒，",
        "有些无聊，",
        "情节紧凑，",
        "表现一般，",
        "令人惊喜，",
        "有点复杂，",
        "节奏缓慢，",
        "特效出色，",
        "情感深沉，",
        "有些许创意，",
        "剧情发展令人陷入其中，",
        "影片的视觉效果让人印象深刻，",
        "角色的发展很出色，",
        "导演的创意发挥得淋漓尽致，",
        "故事情节虽然复杂，但吸引人，"
    ]

    endings = [
        "非常值得一看。",
        "演员表现出色。",
        "导演技巧娴熟。",
        "音效和画面都很出色。",
        "可能不适合所有人。",
        "有很多值得思考的地方。",
        "略显平淡，但总体还行。",
        "我会推荐给我的朋友。",
        "有些地方令人失望。",
        "情节有些许遗憾。"
        "总的来说，这是一部不可错过的佳作。",
        "演员的出色表现使得影片更加精彩。",
        "导演运用了出色的技巧，让整个作品生动有趣。",
        "音效和画面质量都是一流的。",
        "虽然不是每个人的菜，但对于特定观众来说，这是一部精彩的作品。"
    ]


    comment = random.choice(beginningss) + random.choice(beginnings) + random.choice(middles) + random.choice(endings)
    return comment



evaluation = []
user_target_pairs = set()  

for _ in range(30000): 
    while True:
        user_id = random.randint(100001, 150000)
        target_id = random.randint(0, 131035)
        user_target_pair = (1, user_id, target_id)
        
        if user_target_pair not in user_target_pairs:
            break
    
    data = {
        "type_id": 1,
        "target_id": target_id,
        "user_id": user_id,
        "score": generate_random_score(),
        "comment": generate_random_comment()
    }
    
    evaluation.append(data)
    user_target_pairs.add(user_target_pair)

for _ in range(30000): 
    while True:
        user_id = random.randint(100001, 150000)
        target_id = random.randint(0, 9848)
        user_target_pair = (2, user_id, target_id)
        
        if user_target_pair not in user_target_pairs:
            break
    
    data = {
        "type_id": 2,
        "target_id": target_id,
        "user_id": user_id,
        "score": generate_random_score(),
        "comment": generate_random_comment()
    }
    
    evaluation.append(data)
    user_target_pairs.add(user_target_pair)
    
for _ in range(30000): 
    while True:
        user_id = random.randint(100001, 150000)
        target_id = random.randint(0, 10268)
        user_target_pair = (3, user_id, target_id)
        
        if user_target_pair not in user_target_pairs:
            break
    
    data = {
        "type_id": 3,
        "target_id": target_id,
        "user_id": user_id,
        "score": generate_random_score(),
        "comment": generate_random_comment()
    }
    
    evaluation.append(data)
    user_target_pairs.add(user_target_pair)
    
for _ in range(30000):  
    while True:
        user_id = random.randint(100001, 150000)
        target_id = random.randint(0, 12634)
        user_target_pair = (4, user_id, target_id)
        
        if user_target_pair not in user_target_pairs:
            break
    
    data = {
        "type_id": 4,
        "target_id": target_id,
        "user_id": user_id,
        "score": generate_random_score(),
        "comment": generate_random_comment()
    }
    
    evaluation.append(data)
    user_target_pairs.add(user_target_pair)
    
with open('evaluation.json', 'w') as file:
    json.dump(evaluation, file, indent=4)

print("JSON文件已创建")
```
