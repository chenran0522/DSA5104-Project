### TV

```python
import pandas as pd
import os

# Set the file directory or path where the CSV files are located
directory = 'TV series/'

# List to hold all the individual DataFrames
all_dfs = []
filelist = os.listdir(directory)
# Iterate over all the CSV files in the specified directory
for filename in filelist:
    if filename.endswith(".csv"):
        filepath = directory + filename
        print(filepath)
        df = pd.read_csv(filepath)
        all_dfs.append(df)

# Concatenate all the individual DataFrames into one DataFrame
combined_df = pd.concat(all_dfs, ignore_index=True)

# Write the combined DataFrame to a new CSV file
combined_df.to_csv('tv_series.csv', index=False)

print("CSV files have been combined into combined.csv")
```

```python
def process_cast(cast):
    if pd.notna(cast) and'|' in cast:  # Check if the delimiter is present
        parts = cast.split('|')
        pre = parts[0].replace('Director:', '').replace('Directors:', '').replace('Stars:', '').replace('Star:', '').split(',')
        aft = parts[1].replace('Stars:', '').replace('Star:', '').split(',') if len(parts) > 1 else []  # Check if stars information is present
        pre = [e.strip() for e in pre] 
        aft = [e.strip() for e in aft] # Clean up whitespace
        cast_list = pre + aft  # Combine director and stars into one list
        for c in cast_list:
            if c.isspace() or not c:
                cast_list.remove(c)
        return cast_list
    elif pd.notna(cast):
        cast = cast.replace('Stars:', '').replace('Star:', '').replace('Director:', '').replace('Directors:', '').split(',')
        for c in cast:
            if c.isspace() or not c:
                cast.remove(c)
        return [e.strip() for e in cast]
    else:
        return []
```

```python
import re

def extract_runtime(runtime):
    if pd.notna(runtime):  # Check if runtime is not NaN
        match = re.search(r'\d+', str(runtime))  # Convert runtime to string before applying regex
        if match:
            return int(match.group())
    return np.random.randint(80, 180)
```

```python
df = pd.read_csv('tv_series.csv')
df = df.drop_duplicates()
tv = df
# 生成ID
tv['id'] = [i for i in range(7495, 7495 + len(df))]
tv['title'] = df['Title']
tv['rating'] = df['Rating']


tv['watch_numbers'] = np.random.randint(1000, 10000000, size=(len(df)))

tv['runtime'] = df['Runtime'].apply(extract_runtime)
#movie['cast'] = df['Director'] + ', ' + df['Cast']
tv['cast'] = df['Cast'].apply(process_cast)

# Convert Genres to an array
tv['genres'] = df['Genre'].str.split(', ')

# Drop the Director column
#movie.drop(columns=['Director'], inplace=True)

tv = tv[['id', 'title', 'rating','cast','genres', 'watch_numbers', 'runtime']]

tv['format'] = "TV Series"
tv['type_id'] = 1
tv
```

```python
tv.to_json('tv.json', orient='records', lines=True)
```
