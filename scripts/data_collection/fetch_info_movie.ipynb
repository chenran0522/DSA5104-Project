```python
import requests
import time
import csv

API_KEY = 'cceadc0f92feaee38104ac467e4afe75'
BASE_URL = 'https://api.themoviedb.org/3/'

def get_genres():
    endpoint = f'{BASE_URL}genre/movie/list?api_key={API_KEY}'
    response = requests.get(endpoint)
    if response.status_code == 200:
        genres = {genre['id']: genre['name'] for genre in response.json().get('genres', [])}
        return genres
    else:
        print("Error fetching genres:", response.status_code)
        return {}

def get_movie_details(movie_id, genres):
    endpoint = f'{BASE_URL}movie/{movie_id}?api_key={API_KEY}&append_to_response=credits'
    response = requests.get(endpoint)

    if response.status_code == 200:
        movie_data = response.json()
        genre_ids = [genre['id'] for genre in movie_data.get('genres', [])]
        movie_data['genre_names'] = [genres[genre_id] for genre_id in genre_ids if genre_id in genres]
        return movie_data
    else:
        print(f"Error fetching details for movie_id: {movie_id}. Status code: {response.status_code}")
        return None


def get_popular_movies(page=1):
    endpoint = f'{BASE_URL}movie/popular?api_key={API_KEY}&page={page}'
    response = requests.get(endpoint)

    if response.status_code == 200:
        return response.json().get('results', [])
    else:
        print("Error:", response.status_code)
        return []

if __name__ == "__main__":
    genres = get_genres()

    with open('movies.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Title', 'Release Date', 'Overview', 'Director', 'Cast', 'Genres']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        movies_collected = 0
        page = 1

        while movies_collected < 15000:
            movies = get_popular_movies(page)
            
            for movie in movies:
                movie_details = get_movie_details(movie['id'], genres)
                if movie_details:
                    director = next((crew['name'] for crew in movie_details['credits']['crew'] if crew['job'] == 'Director'), 'N/A')
                    cast = [actor['name'] for actor in movie_details['credits']['cast']][:5]  # Get top 5 actors

                    writer.writerow({
                        'Title': movie_details['title'],
                        'Release Date': movie_details.get('release_date', 'N/A'),
                        'Overview': movie_details.get('overview', 'N/A'),
                        'Director': director,
                        'Cast': ', '.join(cast),
                        'Genres': ', '.join(movie_details['genre_names'])
                    })

                    movies_collected += 1
                    if movies_collected >= 10000:
                        break

                # Sleep for a short period to ensure we don't hit the rate limit
                time.sleep(0.05)
            page += 1
```
