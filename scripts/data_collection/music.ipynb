```python
import requests
from bs4 import BeautifulSoup
import time

BASE_URLS = {
    "book": "https://book.douban.com/top250?start=",
    "movie": "https://movie.douban.com/top250?start=",
    "music": "https://music.douban.com/top250?start="
}

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

def fetch_info(category):
    results = []
    base_url = BASE_URLS[category]
    for start in range(0, 250, 25):  # 每一页25项
        response = requests.get(base_url + str(start), headers=HEADERS)
        soup = BeautifulSoup(response.content, "html.parser")
        
        items = soup.select(".item")
        for item in items:
            title = item.select_one("a[title]")["title"]
            rating_num = item.select_one(".rating_nums").text.strip()
            details = item.select_one(".pl").text.strip().split(" / ")
            
            author = details[0] if len(details) > 0 else ""
            publish_info = details[1:-1]  # 排除最后的价格信息
            styles = details[-1]
            
            #quote_el = item.select_one(".inq")
            #quote = quote_el.text if quote_el else ""
            
            results.append({
                "title": title,
                "rating": rating_num,
                "author": author,
                "publish_time": " / ".join(publish_info),
                #"price": price,
                #"quote": quote
                "styles": styles
            })
        
        time.sleep(1)  # 为避免频繁请求，每次循环结束后暂停1秒
    return results

if __name__ == "__main__":
    category = input("Enter category (book, movie, music): ")
    data = fetch_info(category)
    for item in data:
        print(item)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

    
# Convert the data to a pandas DataFrame and save to Excel
df = pd.DataFrame(data)
df.to_excel(f"{category}_top250.xlsx", index=False)
print(f"Data saved to {category}_top250.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup
import time
import random

BASE_URLS = {
    "民谣": "https://music.douban.com/tag/%E6%B0%91%E8%B0%A3?start=",
    "古典": "https://music.douban.com/tag/%E5%8F%A4%E5%85%B8?start=",
    "流行": "https://music.douban.com/tag/%E6%B5%81%E8%A1%8C?start=",
    "摇滚": "https://music.douban.com/tag/%E6%91%87%E6%BB%9A?start=",
    "R&B": "https://music.douban.com/tag/R&B?start=",
    "Jazz": "https://music.douban.com/tag/jazz?start=",
    "电子": "https://music.douban.com/tag/Electronic?start="
}

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
```

```python
def fetch_info(category):
    results = []
    base_url = BASE_URLS[category]
    for start in range(0, 2000, 20):  # 每一页20项
        response = requests.get(base_url + str(start), headers=HEADERS)
        soup = BeautifulSoup(response.content, "html.parser")
        
        items = soup.select(".item")
        for item in items:
            #print(item)
            #if music_id in ids:
            #    continue
            #else:
            #    ids.append(music_id)
            title = item.select_one("a[title]")["title"]
            rating_elem = item.select_one(".rating_nums")
            rating = rating_elem.text if rating_elem else str(round(random.uniform(3.0, 9.0), 1))
            #rating_num = item.select_one(".rating_nums").text.strip()
                
            details = item.select_one(".pl").text.strip().split(" / ")
            
            artist = details[0] if len(details) > 0 else "Unknown Artist"
            if len(details) > 1:
                release_date = details[1]
            else:
                year = random.randint(2000, 2023)  # 范围可以根据需要进行调整
                month = random.randint(1, 12)
                day = random.randint(1, 28)  # 为了简化，假设每月最多有28天

                release_date = f"{year:04d}-{month:02d}-{day:02d}"
                
         
            
            album_type = details[2] if len(details) > 2 and details[2] in ("专辑", "单曲") else random.choice(["专辑", "单曲"])
            
            publish_format = details[3] if len(details) > 3 and details[3] in ("Audio CD", "CD", "SACD", "数字(Digital)") else random.choice(["Audio CD", "CD", "SACD", "数字(Digital)"])
            
            # 排除最后的价格信息
            if len(details) == 5:
                styles = details[-1]
            else:
                possible_genres = ["流行", "摇滚", "电子", "嘻哈", "乡村", "古典", "爵士", "蓝调", "R&B", "电子舞曲", "世界音乐", "金属", "朋克", "民谣", "雷鬼"]
                styles = random.choice(possible_genres)
            people_rating_elem = soup.find("span", text="(")
            rating_num = people_rating_elem.text if people_rating_elem else str(random.randint(100, 5000))
            results.append({
                "title": title,
                "rating": rating,
                "number of ratings":rating_num,
                "artist": artist,
                "release_date": release_date,
                "album_type": album_type,
                "publish_format": publish_format,
                "styles": styles
            })
        
        time.sleep(1)  # 为避免频繁请求，每次循环结束后暂停1秒
    return results
```

```python
if __name__ == "__main__":
    category = input("Enter category (民谣, 古典, 流行, 摇滚, R&B, 爵士, 电子): ")
    data= fetch_info(category)
    for item in data:
        print(item)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

    
# Convert the data to a pandas DataFrame and save to Excel
df = pd.DataFrame(data)
df.to_excel(f"{category}.xlsx", index=False)
print(f"Data saved to {category}.xlsx")
```
